```python
import random
import matplotlib.ticker as ticker
from matplotlib import rcParams
from IPython.display import display, Math, Latex
import matplotlib.pyplot as plt
from math import gcd as bltin_gcd
import numba

@numba.njit
def check_if_prime(n):
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


@numba.njit
def check_statesment(a):
    i = a - 1
    while i > a ** 0.5:
        if a % i == 0 and check_if_prime(i):
            return True
        i -= 1
    return False

count = 0
N = 1000
x = []
y = []
tries = 10000

for i in range(tries):
    x.append(i+1)
    a = random.randint(2, 1_000_000)
    if check_statesment(a):
        count += 1
    y.append(count / (i+1))
print(count/tries)

fig=plt.figure()
plt.tick_params(labelsize = 20)
plt.grid(b=True, color='DarkTurquoise', alpha=0.75, linestyle=':', linewidth=1)
plt.plot(x,y,color='b',lw=2)
plt.plot(x,[0.73166]*len(x),color='r',lw=2)
plt.show()
```

```python
import random
import matplotlib.ticker as ticker
from matplotlib import rcParams
from IPython.display import display, Math, Latex
import matplotlib.pyplot as plt
from math import gcd as bltin_gcd
import numba

@numba.njit
def check_if_triangle(a, b, c):
    return a + b >= c and a + c >= b and c + b >= c


def check_sharp_triangle(a, b, c):
    l = sorted([a, b, c])
    print(sum(l))
    return check_if_triangle(a, b, c) and l[-1] ** 2 < l[0] **  2 + l[1] **  2
    

count = 0
N = 1000
x = []
y = []
tries = 10

for i in range(tries):
    x.append(i+1)
    x1 = random.uniform(0, 1)
    x2 = random.uniform(0, 1)
    if x2 < x1:
        x1, x2 = x2, x1
    a = x1
    b = x2 - x1
    c = 1 - x2
    if check_sharp_triangle(a, b, c):
        count += 1
    y.append(count / (i+1))
print(count/tries)

fig=plt.figure()
plt.tick_params(labelsize = 20)
plt.grid(b=True, color='DarkTurquoise', alpha=0.75, linestyle=':', linewidth=1)
plt.plot(x,y,color='b',lw=2)
plt.plot(x,[0.0397208]*len(x),color='r',lw=2)
plt.show()
```
